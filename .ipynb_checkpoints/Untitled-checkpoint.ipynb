```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
%matplotlib inline


df = pd.read_csv('./survey_results_public.csv')
df
```

```python
num_rows = df.shape[0] #Provide the number of rows in the dataset
num_cols = df.shape[1] #Provide the number of columns in the dataset

print(num_rows, num_cols)
print(df.isnull()['CareerSatisfaction'].mean())
```

```python
count = 0
removed_cols = []
for i in list(df):
    if(df.isnull()[str(i)].mean() > .75):
        removed_cols.append(str(i))
        count = count + 1
print(removed_cols)
df.drop(removed_cols,inplace=True, axis = 1)
df 
```

```python
nanlist = df.columns[df.isna().any()].tolist()
#print(df[['MajorUndergrad']].mode())
#for index, row in df[['MajorUndergrad']].mode().iterrows():
#    if(index == 0):
#        for b in row:
#            print(b)
k = 0
for i in nanlist: 
    for index, row in df[[str(i)]].mode().iterrows():
        if(index == 0):
            for b in row:
                k = b
                df[str(i)] = df[str(i)].fillna(k)
df
```

```python
le = LabelEncoder()
label = le.fit_transform(df['Respondent'])
```

```python

X = []
y = []
for i in list(df):
    if(str(i) == 'CareerSatisfaction'):
        y.append(str(i))
    else:
        X.append(str(i))
```

```python
le = LabelEncoder()
for i in X:
    label = le.fit_transform(df[str(i)])
    df[str(i)] = label
df
```

```python
from sklearn import svm

y = df[['CareerSatisfaction']]
#print(y)
X_train, X_test, y_train, y_test = train_test_split(df[X], y, test_size = .30, random_state=42) 

#lm_model = LinearRegression(normalize=True) # Instantiate
#lm_model.fit(X_train, y_train) #Fit
        
clf = svm.SVC(decision_function_shape='ovo', gamma = 'scale')
clf.fit(X_train, np.ravel(y_train))    
#Predict and score the model
#y_test_preds = lm_model.predict(X_test)
correct_count = 0
for x in range(1, X_test.shape[0]+1):
    if clf.predict(X_test[-x].reshape(1,-1))[0] == y_test[-x]:
        correct_count += 1
print("Accuracy:", (correct_count/X_test.shape[0]) * 100.00)
```
